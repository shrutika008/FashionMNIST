## Feature Scaling

```python
X_train = X_train/255
```

## Split Dataset

```python
from sklearn.model_selection import train_test_split
X_train, x_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.2, random_state = 2020)
```

```python
X_train.shape, y_train.shape
```

```python
x_val.shape, y_val.shape
```

# Convolutional Neural Netwok Model

```python
#keras.layers.conv2D = all convolutional layers
#filters X kernels
#strides = the shift of the our matrix
#padding = adding extra column
#activation
#input = image shape
#channel(3 or single) = single(black and white), 3channel(red,blue,green)
#maxpooling
#pool_size = 2,2
#flatten
#Dense layer - to train the neural netwok(128 = nuerons)
#softmax = to find probability
```

```python
cnn_model = keras.models.Sequential([
                         keras.layers.Conv2D(filters=32, kernel_size=3, strides=(1,1), padding='valid',activation= 'relu', input_shape=[28,28,1]),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Flatten(),
                         keras.layers.Dense(units=128, activation='relu'),
                         keras.layers.Dense(units=10, activation='softmax')
])
```

```python
cnn_model.summary()
```

```python
cnn_model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
#verbos = 1 is progress bar
cnn_model.fit(X_train, y_train, epochs = 10, batch_size = 512, verbose = 1, validation_data = (x_val, y_val))
```

## Model Testing

```python
cnn_model.predict(np.expand_dims(X_test[0], axis = 0)).round(2)
```

```python
#ninth class predicted value
np.argmax(cnn_model.predict(np.expand_dims(X_test[0], axis = 0)).round(2))
```

# Visualization

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,16))
 
j=1
for i in np.random.randint(0, 1000,25):
  plt.subplot(5,5, j); j+=1
  plt.imshow(X_test[i].reshape(28,28), cmap = 'Greys')
  plt.title('Actual = {} / {} \nPredicted = {} / {}'.format(class_labels[y_test[i]], y_test[i], class_labels[np.argmax(y_pred[i])],np.argmax(y_pred[i])))
  plt.axis('off')
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
 
plt.figure(figsize=(16,9))
y_pred_labels = [ np.argmax(label) for label in y_pred ]
cm = confusion_matrix(y_test, y_pred_labels) 
sns.heatmap(cm, annot=True, fmt='d',xticklabels=class_labels, yticklabels=class_labels)
```

```python
#Precison = TP/(TP+FP)
#Recall = TP/(TP+FN)
from sklearn.metrics import classification_report
cr= classification_report(y_test, y_pred_labels, target_names=class_labels)
print(cr)
```

# Complex CNN to improve Accuracy

```python
#dropout = model avoids over fitting
cnn_model2 = keras.models.Sequential([
                         keras.layers.Conv2D(filters=32, kernel_size=3, strides=(1,1), padding='valid',activation= 'relu', input_shape=[28,28,1]),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Conv2D(filters=64, kernel_size=3, strides=(2,2), padding='same', activation='relu'),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Flatten(),
                         keras.layers.Dense(units=128, activation='relu'),        #input layer
                         keras.layers.Dropout(0.25),
                         keras.layers.Dense(units=256, activation='relu'),        #hidden layer
                         keras.layers.Dropout(0.25),
                         keras.layers.Dense(units=128, activation='relu'),        #hidden layer
                         keras.layers.Dense(units=10, activation='softmax')        #output layer
                         ])
```

```python
cnn_model2.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
cnn_model2.fit(X_train, y_train, epochs = 35, batch_size = 512, verbose = 1, validation_data = (x_val, y_val))
```
